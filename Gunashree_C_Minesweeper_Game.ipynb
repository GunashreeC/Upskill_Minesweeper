Minesweeper Game!!

```python
import random

def GenerateMineSweeperMap(n,k):
    arr = [[0 for row in range(n)] for column in range(n)]
    for num in range(k):
        x = random.randint(0,n-1)
        y = random.randint(0,n-1)
        arr[x][y] = 'X'
        if (x >= 0 and x <= n-1)  and (y >= 0 and y <= n-2):
            if arr[x][y+1] != 'X':
                arr[x][y+1] += 1
        if (x >= 0 and x <= n-1)  and (y >= 1 and y <= n-1):
            if arr[x][y-1] != 'X':
                arr[x][y-1] += 1
        if (x >= 1 and x <= n-1)  and (y >= 1 and y <= n-1):
            if arr[x][y-1] != 'X':
                arr[x][y-1] += 1
        if (x >= 1 and x <= n-1)  and (y >= 0 and y <= n-2):
            if arr[x-1][y+1] != 'X':
                arr[x-1][y+1] += 1
        if (x >= 1 and x <= n-1)  and (y >= 0 and y <= n-2):
            if arr[x-1][y] != 'X':
                arr[x-1][y] += 1
        if (x >= 0 and x <= n-2)  and (y >= 0 and y <= n-2):
            if arr[x+1][y+1] != 'X':
                arr[x+1][y+1] += 1
        if (x >= 0 and x <= n-2)  and (y >= 1 and y <= n-1):
            if arr[x+1][y-1] != 'X':
                arr[x+1][y-1] += 1
        if (x >= 0 and x <= n-2)  and (y >= 0 and y <= n-1):
            if arr[x+1][y] != 'X':
                arr[x+1][y] += 1
    return arr

def GeneratePlayerMap(n):
    arr = [['-' for row in range(n)] for coulmn in range(n)]
    return arr

def DisplayMap(map):
    for row in map:
        print(" ".join(str(cell) for cell in row))
        print("")
        
def CheckPlaying(map):
    for row in map:
        for cell in row:
            if cell =='-':
                return True
    return False

def CheckContinueGame(score):
    print("Your score: ",score)
    iscontinue = input('Do you want to play again?(y/n): ')
    if iscontinue == 'n':
        return False
    
    return True

def Game():
    GameStatus = True
    while GameStatus:
        difficulty = input('Select your difficulty(b, i, h): ')
        if difficulty.lower() == 'b':
            n = 5
            k = 3
        elif difficulty.lower() == 'i':
            n = 6
            k = 8
        else:
            n = 8
            k = 20
            
        minesweeper_map = GenerateMineSweeperMap(n,k)
        player_map = GeneratePlayerMap(n)
        score = 0
        
        while True:
            if CheckPlaying(player_map):
                choice = int(input('1.Open a cell?\n2.Flag a bomb? '))
                if choice == 1:
                    print ('Enter the cell you want to open :')
                    x = input('X (1 to 5):')
                    y = input('Y (1 to 5):')
                    x = int(x) - 1
                    y = int(y) - 1
                    if (minesweeper_map[x][y] == 'X'):
                        print("Game over")
                        DisplayMap(minesweeper_map)
                        GameStatus = CheckContinueGame(score)
                        break

                    else:
                        player_map[x][y] = minesweeper_map[x][y]
                        DisplayMap(player_map)
                        score +=1
                else:
                    print("Enter the cell you want to flag: ")
                    x = input("X(1 to 5) : ")
                    y = input("Y(1 to 5) : ")
                    x = int(x) - 1
                    y = int(y) - 1
                    player_map[x][y] = 'X'
                    DisplayMap(player_map)

            elif player_map == minesweeper_map:
                DisplayMap(player_map)
                print("You won!!")
                GameStatus = CheckContinueGame(score)
                break

            else:
                print("Game Over!")
                DisplayMap(minesweeper_map)
                GameStatus = CheckContinueGame(score)
                break
```

```python
Game()
```
